```python
import numpy as np
import math

# Given values
a = 1 / math.sqrt(2)
b = 1 / math.sqrt(2)

# Create vector
v = np.array([a, b])
print("Vector:", v)

# Check normalization
norm_check = a**2 + b**2
print("a^2 + b^2 =", norm_check)

if math.isclose(norm_check, 1.0):
    print("Vector is normalized")
else:
    print("Vector is NOT normalized")

# Length of vector
length = np.linalg.norm(v)
print("Length of vector:", length)
```

```python
import numpy as np

# Read vector from user
n = int(input("Enter size of vector: "))
vec = []

for i in range(n):
    val = float(input(f"Enter element {i+1}: "))
    vec.append(val)

A = np.array(vec)
print("Vector A:", A)

# Function: check normalization
def is_normalized(v):
    return np.isclose(np.sum(v**2), 1)

print("Is normalized?", is_normalized(A))

# Function: dot and cross product
def dot_cross(A, B):
    dot = np.dot(A, B)
    cross = np.cross(A, B)
    return dot, cross

# Input second vector
vec2 = []
for i in range(n):
    val = float(input(f"Enter element {i+1} of vector B: "))
    vec2.append(val)

B = np.array(vec2)

d, c = dot_cross(A, B)
print("Dot product:", d)
print("Cross product:", c)

# Function: unit vector
def unit_vector(v):
    mag = np.linalg.norm(v)
    if mag == 0:
        return "Zero vector has no unit vector"
    return v / mag

print("Unit vector of A:", unit_vector(A))
```

```python
import numpy as np

r = int(input("Enter rows: "))
c = int(input("Enter columns: "))

mat = []
print("Enter matrix elements row-wise:")
for i in range(r):
    row = []
    for j in range(c):
        row.append(float(input()))
    mat.append(row)

A = np.array(mat)
print("Matrix A:\n", A)

# Transpose
print("Transpose:\n", A.T)

# Determinant (only square)
if r == c:
    print("Determinant:", np.linalg.det(A))

    # Rank
    print("Rank:", np.linalg.matrix_rank(A))

    # Trace
    print("Trace:", np.trace(A))

    # Eigenvalues & Eigenvectors
    vals, vecs = np.linalg.eig(A)
    print("Eigenvalues:\n", vals)
    print("Eigenvectors:\n", vecs)

    # Inverse
    if np.linalg.det(A) != 0:
        print("Inverse:\n", np.linalg.inv(A))
    else:
        print("Matrix is singular, inverse not possible")
else:
    print("Eigen, determinant, inverse only for square matrices")
```

```python
# Read complex number
z = complex(input("Enter a complex number (e.g. 3+4j): "))
print("Complex number:", z)

# Conjugate
print("Conjugate:", z.conjugate())

# Second number
z2 = complex(input("Enter another complex number: "))

print("Addition:", z + z2)
print("Subtraction:", z - z2)
print("Multiplication:", z * z2)
print("Division:", z / z2)

# As 2D vector
print("As 2D vector: (", z.real, ",", z.imag, ")")

# Magnitude
print("Magnitude:", abs(z))
```
